<center><span style="color: red;">Importamos las Principales Librerias a utilizar</span></center>

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import numpy as np
from scipy import stats
```

<center><span style="color: red;">Creamos un DataFrame con los datos de  todas las empresas del SP500.</span></center>
<center><span style="color: red;">Creamos un csv apartir del mismo.</span></center>

```python


df = pd.DataFrame()
for ticker in sp500 :
    data= yf.download(ticker, start="2000-01-01", end="2023-03-21")
    
    data.insert(0, 'Empresa', ticker)
    data.reset_index(inplace=True)
    data['Date'] = pd.to_datetime(data['Date'])

    # Agregar columnas con el año, mes y día de la fecha
    data['Year'] = data['Date'].dt.year
    data['Month'] = data['Date'].dt.month
    data['Day'] = data['Date'].dt.day
    
    df = df.append(data)

df.to_csv('SP500.csv')
```

<center><span style="color: red;">Creamos una lista de las empresas que se pudieron</span></center>
<center><span style="color: red;">descargar para poder hacer el analisis</span></center>

```python
sp500_unicos=df['Empresa'].str.strip().unique()
```

<center><span style="color: red;">Creamos una DataFrame para poder elegir las 10 empresas</span></center>
<center><span style="color: red;">de las cuales se realizara el analis de datos</span></center>

```python

df_conclusion=pd.DataFrame(columns=['Empresa','Promedio_Volumen','sum_Open_Close'])
for sp500 in sp500_unicos:
    dftest=pd.read_csv('SP500.csv')
    dftest=dftest[dftest['Empresa']==sp500]
    dftest['Open_Close'] = dftest['Close'] - dftest['Open']
    nueva_fila = [sp500, dftest['Volume'].mean(), dftest['Open_Close'].sum()]
    df_conclusion=df_conclusion.append(pd.Series(nueva_fila, index=df_conclusion  .columns), ignore_index=True)

df_conclusion
```

<center><span style="color: orange;">Método de intercuartil  </span></center>
<center><span style="color: orange;">El método de intercuartil se basa en el rango intercuartil (IQR) para detectar valores atípicos. </span></center>
<center><span style="color: red;">Los valores atípicos se identifican como aquellos puntos de datos que están por debajo de Q1 - 1.5 * IQR o por encima de Q3 + 1.5 * IQR. </span></center>

```python
for ticker in sp500_para_evaluar:
    
    df_out=df_outliers[df_outliers['Empresa']==ticker]
    print(ticker)
    for col in columnas:
        print(col)
        Q1 = df_out[col].quantile(0.25)
        Q3 = df_out[col].quantile(0.75)
        IQR = Q3 - Q1
        outliers = df_out[(df_out[col] < Q1 - 1.5 * IQR) | (df_out[col] > Q3 + 1.5 * IQR)]
        print(outliers)
```
